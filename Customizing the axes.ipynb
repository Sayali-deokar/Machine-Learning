# Customizing the axes

Importing Library as follows

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

Importing dataset form local host

```python
iris=pd.read_csv("F:/Datasets/iris.csv")
```

Customizing the axes Ploting

```python
plt.scatter(iris['Sepal.Length'], iris['Sepal.Width'])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title('Scatter Plot')
plt.axis((4,9,1,5))
plt.show()
```

```python
plt.subplot(1,2,1)
plt.hist(iris['Sepal.Length'],color='red')
plt.title("Sepal Length of Iris")

plt.subplot(1,2,2)
plt.hist(iris['Sepal.Width'])
plt.title("Sepal Width of Iris")

plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,2,1)
plt.hist(iris['Sepal.Length'],color='red')
plt.title("Sepal Length of Iris")

plt.subplot(2,2,2)
plt.hist(iris['Sepal.Width'])
plt.title("Sepal Width of Iris")

plt.subplot(2,2,3)
plt.hist(iris['Petal.Length'],color='red')
plt.title("Petal Length of Iris")

plt.subplot(2,2,4)
plt.hist(iris['Petal.Width'])
plt.title("Petal Width of Iris")

plt.tight_layout()
plt.show()
```
