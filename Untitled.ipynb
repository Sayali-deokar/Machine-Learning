# Using Pandas

Importing Library as follows

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

Importing dataset form local host

```python
iris=pd.read_csv("F:/Datasets/iris.csv")
```

Hist Ploting

```python
iris['Sepal.Length'].plot(kind='hist')
plt.show()

iris['Sepal.Length'].plot(kind='hist',color="red",bins=10)
plt.show()

iris['Sepal.Length'].plot(kind='hist',color="red",
                          alpha=0.25, bins=20)
plt.show()

iris['Sepal.Length'].plot(kind='hist',color="purple",
    alpha=0.4,bins=20)
plt.xlabel('Sepal Length')
plt.ylabel('Count')
plt.title('Histogram')
plt.show()
```

Importing dataset form local host

```python
Orders = pd.read_csv("F:/Datasets/Orders.csv")
```

Bar Ploting

```python
cts = Orders['Payment Terms'].value_counts()
cts.plot(kind='bar')
plt.show()

cts = Orders['Payment Terms'].value_counts()
cts.plot(kind='bar')
plt.xticks(rotation=0)
plt.show()

cts = Orders['Payment Terms'].value_counts()
cts.plot(kind='bar', color="violet")
plt.show()
```

Pie Ploting

```python
cts.plot(kind='pie')
plt.show() 

cts.plot(kind='pie',autopct="%.2f%%")
plt.show() 
 
cts.plot(kind='pie',autopct="%.2f%%",shadow=True)
plt.show() 
```

Kde Ploting

```python
iris['Sepal.Length'].plot(kind='kde')
plt.show()

iris.plot(kind='scatter', x='Sepal.Length', 
          y='Sepal.Width')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter Plot")
plt.show()
```

Box Ploting

```python
iris['Sepal.Length'].plot(kind='box')
plt.show()

iris['Sepal.Width'].plot(kind='box')
plt.show()

iris['Petal.Length'].plot(kind='box')
plt.show()

iris['Petal.Width'].plot(kind='box')
plt.show()

iris.plot(kind='box')
plt.show()
```
