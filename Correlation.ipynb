# Correlation

Importing Library as follows

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
```

Importing dataset form local host

```python
df=pd.read_csv("D:/Statistics (Python)/Datasets/iris.csv")
```

Importing matplotlib as follows

```python
import matplotlib.pyplot as plt
```

Ploting

```python
plt.scatter(df['Sepal.Length'],df['Petal.Length'])
plt.show()
```

```python
df['Sepal.Length'].corr(df['Petal.Length'])
```

OR

```python
pearsonr(df['Sepal.Length'],df['Petal.Length'])
```

Correlation Matrix

```python
np.corrcoef(df['Sepal.Length'],df['Petal.Length'])
```

OR

```python
df.corr()
```

 Scatter Matrix Plot

```python
import seaborn as sns
sns.pairplot(df)
plt.show()
```

Heatmap of Correlation Matrix

```python
df2 = df.drop(["Species"],axis="columns")
plt.matshow(df2.corr())
plt.xticks(range(len(df2.columns)), df2.columns,rotation=45)
plt.yticks(range(len(df2.columns)), df2.columns)
plt.colorbar()
plt.show()
```

OR

```python
sns.heatmap(df2.corr())
plt.yticks(range(len(df2.columns)), 
           df2.columns,rotation = 45)
plt.show()
```
