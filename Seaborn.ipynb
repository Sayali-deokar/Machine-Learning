# Seaborn

Importing Library as follows

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Importing dataset form local host

```python
iris=pd.read_csv("F:/Datasets/iris.csv")
```

```python
sns.boxplot(y='Sepal.Length', data=iris)
plt.show()

sns.boxplot(x='Species', y='Sepal.Length', data=iris)
plt.show()

sns.distplot(iris['Sepal.Length'])
plt.show()

sns.scatterplot(x='Sepal.Length', y='Sepal.Width', data=iris)
plt.show()

sns.scatterplot(x='Sepal.Length', y='Sepal.Width',
                hue='Species',
                data=iris)
plt.show()

sns.relplot(x='Sepal.Length', y='Sepal.Width',
                hue='Species',
                data=iris)
plt.show()

sns.regplot(x='Sepal.Length', y='Sepal.Width', data=iris)
plt.show()
```

 Facets

```python
g = sns.FacetGrid(iris, col="Species")
g.map(plt.hist, "Sepal.Length")
plt.show()

g = sns.FacetGrid(iris, row="Species")
g.map(plt.hist, "Sepal.Length")
plt.show()

g = sns.FacetGrid(iris, col="Species")
g.map(plt.scatter, "Sepal.Length","Sepal.Width")
plt.show()
```

JointPlot

```python
sns.jointplot(x='Sepal.Length', y='Sepal.Width', data=iris)
plt.show()

sns.jointplot(x='Sepal.Length', y='Sepal.Width', data=iris,kind="reg")
plt.show()
```

Pairs Plot / Matrix Plot

```python
sns.pairplot(iris)
plt.show()

sns.pairplot(iris, hue="Species")
plt.show()
```

 Heatmap

```python
sns.heatmap( iris.corr(),annot=True)
plt.show()

sns.heatmap( iris.corr(),cmap="YlGnBu",annot=True)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```
