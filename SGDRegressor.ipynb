# SGDRegressor

Importing Library as follows

```python
import pandas as pd
import numpy as np
```

Importing dataset form local host

```python
df_concrete = pd.read_csv("D:/Statistics (Python)/Cases/Concrete Strength/Concrete_Data.csv")
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import ElasticNet
X = df_concrete.iloc[:,:8]
y = df_concrete.iloc[:,8]
```

Create training and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, 
                                                    random_state=42)
```

SGD Regressor 

```python
from sklearn.linear_model import SGDRegressor
sgdReg = SGDRegressor(random_state=2019)
sgdReg.fit(X_train,y_train)
y_pred = sgdReg.predict(X_test)
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print(mean_squared_error(y_test, y_pred) ** 0.5)
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```

Model Tunning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
kfold = KFold(n_splits=5, random_state=42)

eta_range = np.linspace(0.001,1.2,10)


lr_range = ['constant','adaptive','invscaling','optimal']
sgdReg = SGDRegressor(random_state=2019,max_iter=500)

parameters = dict( eta0 = eta_range, learning_rate =lr_range)

svmGrid = GridSearchCV(sgdReg, param_grid=parameters, cv=kfold,
                       scoring='r2',verbose=2)
svmGrid.fit(X, y)
```

Best Parameters

```python
print(svmGrid.best_params_)

print(svmGrid.best_score_)
```

random grid

```python
from sklearn.model_selection import RandomizedSearchCV

rndGrid = RandomizedSearchCV(sgdReg,n_iter=100,random_state=2012,
                             param_distributions=parameters,
                             scoring='r2',verbose=2)

rndGrid.fit(X,y)
```

Best Parameters

```python
print(rndGrid.best_params_)

print(rndGrid.best_score_)
```
