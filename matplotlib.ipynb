# Using matplotlib

Importing Library as follows

```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd
```

Ploting parameters

```python
x = [1990,1995,2000,2005,2010,2015]
y = [800,345,125,423,560,782]
```

Scatter Plot

```python
plt.scatter(x,y)
plt.show()
```

 How not to plot

```python
plt.bar(x, y, align='center', alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Sales ('000)")
plt.title('Sales Comparison')
plt.show()
```

How to plot

```python
x = ['1990','1995','2000','2005','2010','2015']
```

OR

```python
x = [str(v) for v in x]
plt.bar(x, y, align='center', alpha=0.5)
plt.xlabel("Year")
plt.ylabel("Sales ('000)")
plt.title('Sales Comparison')
plt.show()
```

How to plot

```python
y_pos = np.arange(len(x))
plt.bar(y_pos, y, align='center', alpha=0.5)
plt.xticks(y_pos, x)
plt.xlabel("Year")
plt.ylabel("Sales ('000)")
plt.title('Sales Comparison')
plt.show()

x = [1990,1995,2000,2005,2010,2015]
y = [800,345,125,423,560,782]
plt.plot(x,y)
plt.xlabel("Year")
plt.ylabel("Sales ('000)")
plt.title('Sales Analysis')
plt.show()

x = [1990,1995,2000,2005,2010,2015]
y = [800,345,125,423,560,782]
plt.scatter(x,y)
plt.show()
```
